```python
import os
import json
from pathlib import Path
from dotenv import load_dotenv

def resolve_workdir() -> Path:
    cwd = Path.cwd()
    candidates = [cwd, cwd.parent, cwd.parent.parent]
    for c in candidates:
        if (c / 'config.env').exists():
            return c
    return cwd

WORKDIR = resolve_workdir()
load_dotenv(WORKDIR / 'config.env')

AZURE_OPENAI_ENDPOINT = os.getenv('AZURE_OPENAI_ENDPOINT', '').strip()
USER_ALIAS = os.getenv('USER_ALIAS', '').strip().lower().replace('_', '-')
USER_RG_NAME = os.getenv('USER_RG_NAME', '').strip()
AZURE_SUBSCRIPTION_ID = os.getenv('AZURE_SUBSCRIPTION_ID', '').strip()

MODEL_DEPLOYMENT_NAME = os.getenv('MODEL_DEPLOYMENT_NAME', 'gpt-4o').strip()
API_VERSION = os.getenv('API_VERSION', '2025-01-01-preview').strip()
AZURE_LOCATION = os.getenv('AZURE_LOCATION', 'eastus').strip()
EMBEDDING_MODEL_DEPLOYMENT = os.getenv('EMBEDDING_MODEL_DEPLOYMENT', 'text-embedding-3-small').strip()

assert AZURE_OPENAI_ENDPOINT and 'openai.azure.com' in AZURE_OPENAI_ENDPOINT, 'Configura AZURE_OPENAI_ENDPOINT'
assert USER_ALIAS, 'Configura USER_ALIAS'
assert USER_RG_NAME, 'Configura USER_RG_NAME'
assert AZURE_SUBSCRIPTION_ID, 'Configura AZURE_SUBSCRIPTION_ID'

print('OK variables de entorno')
print('Workdir:', WORKDIR)
print('Alias:', USER_ALIAS)
print('RG:', USER_RG_NAME)
print('Subscription:', AZURE_SUBSCRIPTION_ID)
```

```python
from azure.identity import DefaultAzureCredential, get_bearer_token_provider
from openai import AzureOpenAI

credential = DefaultAzureCredential()
token_provider = get_bearer_token_provider(credential, 'https://cognitiveservices.azure.com/.default')
client = AzureOpenAI(
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    azure_ad_token_provider=token_provider,
    api_version=API_VERSION,
)

probe = client.models.list()
print('OK autenticacion Azure OpenAI')
_ = probe
```

```python
workshop_config = {
    'azure_openai_endpoint': AZURE_OPENAI_ENDPOINT,
    'model_deployment_name': MODEL_DEPLOYMENT_NAME,
    'api_version': API_VERSION,
    'azure_location': AZURE_LOCATION,
    'embedding_model_deployment': EMBEDDING_MODEL_DEPLOYMENT,
    'user_alias': USER_ALIAS,
    'user_rg_name': USER_RG_NAME,
    'azure_subscription_id': AZURE_SUBSCRIPTION_ID,
    'resource_prefix': f'contoso-{USER_ALIAS}',
    'setup_verified': True
}

config_path = WORKDIR / 'workshop_config.json'
config_path.write_text(json.dumps(workshop_config, indent=2), encoding='utf-8')

print('OK workshop_config.json generado')
print('Path:', config_path)
```
